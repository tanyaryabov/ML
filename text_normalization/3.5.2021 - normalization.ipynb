## Cleaning the classes

```python
import inflect
p = inflect.engine()
```

Let's clean up the `DATE` class. We'll move any numbers that are not 4 digits to the `CARDINAL` class. 

```python
non_year = test.before.apply(len) != 4
test.loc[(test["class"] == "DATE") & (test.before.str.match('^[0-9]+$')) & non_year, "class"] = "CARDINAL"
test[(test["class"] == "DATE") & (test.before.str.match('^[0-9]+$')) & non_year]
```

Let's move those in `DIGIT` to `DATE`, since they're all in the format of `2021-01-01`. 

```python
test.loc[(test["class"] == "DIGIT") & (test["before"].str.contains('-')), "class"] = "DATE"
test[test['class'] == 'DATE']
```

Move some dates from `TELEPHONE` to `DATES`. 

```python
dates1 = test[test["class"] == "TELEPHONE"]["before"]
dates1 = dates1.tolist()

dates2 = [date for date in dates1 if re.findall('^\d{4}-\d{2}-\d{2}$', date)]
dates2

test.loc[(test["class"] == "TELEPHONE") & (test["before"].isin(dates2)), "class"] = "DATE"
test.loc[(test["class"] == "DATE") & (test["before"].isin(dates2))]
```

Let's also move anything that is a number + some characters to measure. For example, `12 m`.

```python
# move measures from DATE to MEASURE
measures1 = test[test["class"] == "DATE"]["before"]
measures1 = measures1.tolist()

measures2 = [m for m in measures1 if re.findall('^[0-9]{1,10} [A-Za-z]{1,2}$', m)]
test.loc[(test["class"] == "DATE") & (test["before"].isin(measures2)), "class"] = "MEASURE"
```

Move telephone values like `-12` to `Cardinal`. 

```python
neg1 = test[test["class"] == "TELEPHONE"]["before"].values

neg2 = [n for n in neg1 if re.findall('^-[0-9]{1,9}$', n)]
neg2
test.loc[(test["class"] == "DATE") & (test["before"].isin(neg2)), "class"] = "CARDINAL"
```

## Normalize `Cardinal`, `Decimal`, `Ordinal`, `Digit`, and `Time`. 

```python
list = ['CARDINAL', 'DECIMAL', 'ORDINAL', 
        'DIGIT', 'TIME']
test.loc[test["class"]
         .isin(list), "after"] = test[test["class"]
         .isin(list)]["before"].apply(p.number_to_words)
                                                        
test.loc[test["class"].isin(['PUNCT', 'PLAIN']),"after"]= test.loc[test["class"].isin(['PUNCT', 'PLAIN']),"before"]
```

## Normalize `letters`

```python
exclude = set(string.punctuation)
# https://www.pythonpool.com/remove-punctuation-python/
test.loc[test["class"] == "LETTERS", "after"] = test[test["class"] == "LETTERS"]["before"].str.lower()
test.loc[test["class"] == "LETTERS", "after"] = test[test["class"] == "LETTERS"]["after"].str.replace(" ", "")
test[test['class'] == "LETTERS"]
s = test[test["class"] == "LETTERS"]["after"]
# " ".join()
letters = []
for i in s:
    letters.append(' '.join(ch for ch in i if ch not in exclude))
    # https://stackoverflow.com/questions/18221436/efficient-way-to-add-spaces-between-characters-in-a-string
test.loc[test["class"] == "LETTERS", "after"] = letters
test[test["class"] == "LETTERS"]
```

## Normalize `telephone`

```python
telephone = []
for i in test[test["class"] == "TELEPHONE"]["before"]:
    i = i.replace("-", ".")
    telephone.append(p.number_to_words(i, group=1, decimal="sil").replace(",", ""))
    
test.loc[test["class"] == "TELEPHONE", "after"] = telephone
test[test["class"] == "TELEPHONE"]

# https://pypi.org/project/inflect/
```

## Normalize `digit`

```python
digit = []
for i in test[test["class"] == "DIGIT"]["before"]:
    digit.append(p.number_to_words(i, group=1).replace(",", ""))
    
test.loc[test["class"] == "DIGIT", "after"] = digit
test[test["class"] == "DIGIT"]

# https://pypi.org/project/inflect/
```

## Normalize `plain`

```python
plain = []
for i in test[test["class"] == "PLAIN"]["before"]:
    if i == "vol":
        i = "volume"
    elif i == 'vs':
        i = "versus"
    elif i == "st":
        i = "saint"
    elif i == "vol":
        i = "volume"
    elif i == "-":
        i = "to"
    plain.append(i)

test.loc[test["class"] == "PLAIN", "after"] = plain
test[(test["class"] == "PLAIN") & (test["before"] != test["after"])]
```

## Normalize `fraction`

```python
fractions1 = test[test["class"] == "FRACTION"]["before"]
fractions2 = [f for f in fractions1 if re.match('^[0-9]{1,9}/[0-9]{1,9}$', f)]
fractions = []
for i in fractions2:
    numerator, denominator = i.split('/')
    numerator = p.number_to_words(numerator, andword="").replace(",", "")  
    if str(denominator)[-2:] == "12": # ends in 12
        denominator = p.number_to_words(denominator, andword="").replace(",", "")
        denominator = " ".join(denominator.split(" ")[:-1])
        end = "twelfths"
        if "hundred" not in denominator.split(" ")[-2:]:
            denominator = end
        else:
            denominator = denominator + " " + end
    elif str(denominator)[-1:] == "2": # ends in 2
        denominator = p.number_to_words(denominator, andword="").replace(",", "")
        end = denominator.split(" ")[-1].replace("two", "halves")
        if "hundred" not in denominator.split(" ")[-2:]:
            denominator = end
        else:
            denominator = " ".join(denominator.split(" ")[:-1])
            denominator = denominator + " " + end
    elif str(denominator)[-2:] == "13": # ends in 13
        denominator = p.number_to_words(denominator, andword="").replace(",", "")
        denominator = " ".join(denominator.split(" ")[:-1])
        end = "thirteenths"
        if "hundred" not in denominator.split(" ")[-2:]:
            denominator = end
        else:
            denominator = denominator + " " + end
    elif str(denominator)[-1:] == "3": # ends in 3
        denominator = p.number_to_words(denominator, andword="").replace(",", "")
        end = denominator.split(" ")[-1].replace("three", "thirds")
        if "hundred" not in denominator.split(" ")[-2:]:
            denominator = end
        else:
            denominator = " ".join(denominator.split(" ")[:-1])
            denominator = denominator + " " + end
    elif str(denominator)[-1:] == "5": # ends in 5
        denominator = p.number_to_words(denominator, andword="").replace(",", "")
        end = denominator.split(" ")[-1].replace("five", "fifths")
        if "hundred" not in denominator.split(" ")[-2:]:
            denominator = end
        else:
            denominator = " ".join(denominator.split(" ")[:-1])
            denominator = denominator + " " + end       
    elif str(denominator)[-1:] == "8": # ends in 8
        denominator = p.number_to_words(denominator, andword="").replace(",", "")
        end = denominator.split(" ")[-1].replace("eight", "eighths")
        if "hundred" not in denominator.split(" ")[-2:]:
            denominator = end
        else:
            denominator = " ".join(denominator.split(" ")[:-1])
            denominator = denominator + " " + end 
    elif str(denominator)[-1:] == "9": # ends in 9
        denominator = p.number_to_words(denominator, andword="").replace(",", "")
        end = denominator.split(" ")[-1].replace("nine", "ninths")
        if "hundred" not in denominator.split(" ")[-2:]:
            denominator = end
        else:
            denominator = " ".join(denominator.split(" ")[:-1])
            denominator = denominator + " " + end 
    elif (int(denominator) % 10 == 0) & (int(denominator) % 100 != 0) & (int(denominator) != 10): # 10s but not 100s
        denominator = p.number_to_words(denominator, andword="").replace(",", "")
        denominator = denominator.replace("y", "ieths")
    else:
        denominator = p.number_to_words(denominator, andword="").replace(",", "")
        end = "ths"
        denominator = denominator + end
    denominator = denominator.replace("  ", " ")
    if numerator == "one":
        i = numerator + " " + denominator[:-1]
        i = i.replace("one halve", "one half").replace("one oneths", "one")
    else:
        i = numerator + " " + denominator
    fractions.append(i)
    
test.loc[(test["class"] == "FRACTION") & (test["before"].isin(fractions2)), "after"] = fractions
test.loc[(test["class"] == "FRACTION") & (test["before"].isin(fractions2))]
```

```python
address = test[test["class"] == "ADDRESS"]["before"].values
addresses = []
for i in address:
    i = i.replace('-', "").replace(" ", "")
    letter = [x for x in i if re.match('[A-Za-z]{1,9}' , x)]
    nums = [x for x in i if re.match('[0-9]{4}', x)]
    letter = (' '.join(ch for ch in letter))
    letter = letter.lower()
    nums = p.number_to_words(i, group=1, zero='o').replace(",", "")
    i = letter + " " + nums
    addresses.append(i)
test.loc[(test["class"] == "ADDRESS") & (test["before"].isin(address)), "after"] = addresses
test.loc[(test["class"] == "ADDRESS") & (test["before"].isin(address))]
```

## Normalize `date`

```python
def date_norm(x):
    
    dates_list = []

    for date in x:
        date = parse(date).strftime("%B %d %Y")
        # https://stackoverflow.com/questions/466345/converting-string-into-datetime
        month, day, year = date.split(" ")


        ordinal = lambda n: "%d%s" % (n,"tsnrhtdd"[(n//10%10!=1)*(n%10<4)*n%10::4])
        # https://stackoverflow.com/questions/9647202/ordinal-numbers-replacement

        day = ordinal(int(day))
        day = p.number_to_words(day)
        month = month.lower()
        if (int(year) % 100 == 0) & (int(year) < 2000):
            year = year.replace("0", "")
            year = p.number_to_words(year).replace(",", "") + " hundred"
        elif int(year) < 2000:
            year = p.number_to_words(year, andword="", group=2, zero="o").replace(",", "")
        else: 
            year = p.number_to_words(year, andword="").replace(",", "")

        date_formatted = "the " + day + " of " + month + " " + year
        dates_list.append(date_formatted)

    test.loc[(test["class"] == "DATE") & (test["before"].isin(dates2)), "after"] = dates_list
```

```python
dates1 = test[test["class"] == "DATE"]["before"]
dates1 = dates1.tolist()

dates2 = [date for date in dates1 if re.findall('^\d{4}-([1-9]|1[0-2])-\d{2}$', date)]
# https://stackoverflow.com/questions/7105146/how-to-represent-regex-number-ranges-e-g-1-to-12
date_norm(dates2)

test.loc[(test["class"] == "DATE") & (test["before"].isin(dates2))]
```

## Normalize `verbatim`

```python
verbatim = test[test["class"] == "VERBATIM"]["before"].values
verbatims = []
for i in verbatim:
    letters = (' '.join(ch for ch in i))
    letters = letters.lower().replace("&", "and").replace("#", "number").replace("$", "dollar").replace(".", "dot")
    verbatims.append(letters)
test.loc[(test["class"] == "VERBATIM") & (test["before"].isin(verbatim)), "after"] = verbatims
test.loc[(test["class"] == "VERBATIM") & (test["before"].isin(verbatim))]
```

## Normalize `address`

```python
address = test[test["class"] == "ADDRESS"]["before"].values
addresses = []
for i in address:
    i = i.replace('-', "").replace(" ", "")
    letter = [x for x in i if re.match('[A-Za-z]{1,9}' , x)]
    nums = [x for x in i if re.match('[0-9]{4}', x)]
    letter = (' '.join(ch for ch in letter))
    letter = letter.lower()
    nums = p.number_to_words(i, group=1, zero='o').replace(",", "")
    i = letter + " " + nums
    addresses.append(i)
test.loc[(test["class"] == "ADDRESS") & (test["before"].isin(address)), "after"] = addresses
test.loc[(test["class"] == "ADDRESS") & (test["before"].isin(address))]
```

## Normalize `electronic`

```python
link = test[test["class"] == "ELECTRONIC"]["before"]
links = []
for i in link:
    i = " ".join(i)
    i = i.replace("/", "slash").replace(":", "colon").replace(".", "dot").replace("colon colon", "::")
    i = i.replace("_", "underscore")
    i = i.lower()
    links.append(i)

test.loc[(test["class"] == "ELECTRONIC") & (test["before"].isin(link)), "after"] = links
test.loc[(test["class"] == "ELECTRONIC") & (test["before"].isin(link))]
```
